```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

sizes = [1,2,4,8,16]
```

```python
fifo = {
    's': [[1,2,4,8,16], [28220, 21583, 18155, 16828, 17346]],
    'sp': [[1,2,4,8,16], [43881, 37910, 35993, 37345, 42914]]
}
plt.scatter(x=fifo[0], y=fifo[1])
plt.ylabel("Number of Page Swaps")
plt.xlabel("Page Size")
plt.title("Number of Page Swaps using FIFO without Pre-paging")
plt.show()
```

```python
fifo_pre = [[1,2,4,8,16], [43881, 37910, 35993, 37345, 42914]]
plt.scatter(x=fifo_pre[0], y=fifo_pre[1])
plt.ylabel("Number of Page Swaps")
plt.xlabel("Page Size")
plt.title("Number of Page Swaps using FIFO with Pre-paging")
plt.show()
```

```python
lru = [[1,2,4,8,16], [17839, 11997, 10450, 11454, 10333]]
plt.scatter(x=lru[0], y=lru[1])
plt.ylabel("Number of Page Swaps")
plt.xlabel("Page Size")
plt.title("Number of Page Swaps using LRU without Pre-paging")
plt.show()
```

```python
lru_pre = [[1,2,4,8,16], [98683, 38066, 208271, 216279, 370469]]
plt.scatter(x=lru_pre[0], y=lru_pre[1])
plt.ylabel("Number of Page Swaps")
plt.xlabel("Page Size")
plt.title("Number of Page Swaps using LRU without Pre-paging")
plt.show()
```

```python
clock = {
    's': [2375, 1307, 1216, 1033, 2492],
    'sp': [3573, 1450, 754, 542, 162],
    'f': [2375, 1307, 1216, 1033, 2492],
    'fp': [1806, 732, 382, 273, 81]
}
plt.figure(figsize=(12,12))

plt.subplot(221)
plt.scatter(x=sizes, y=clock['s'])
plt.ylabel("Number of Page Swaps")
plt.xlabel("Page Size")
plt.title("Number of Page Swaps using Clock without Pre-paging")

plt.subplot(222)
plt.scatter(x=sizes, y=clock['sp'])
plt.ylabel("Number of Page Swaps")
plt.xlabel("Page Size")
plt.title("Number of Page Swaps using Clock with Pre-paging")

plt.subplot(223)
plt.scatter(x=sizes, y=clock['f'])
plt.ylabel("Number of Page Faults")
plt.xlabel("Page Size")
plt.title("Number of Page Faults using Clock without Pre-paging")

plt.subplot(224)
plt.scatter(x=sizes, y=clock['fp'])
plt.ylabel("Number of Page Faults")
plt.xlabel("Page Size")
plt.title("Number of Page Faults using Clock with Pre-paging")

plt.show()
```
